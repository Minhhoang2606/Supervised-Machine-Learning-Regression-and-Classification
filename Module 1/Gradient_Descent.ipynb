```python
import numpy as np
import matplotlib.pyplot as plt

# Define a simple quadratic cost function and its derivative
def cost_function(w):
    return w**2

def gradient(w):
    return 2 * w

# Initialize parameters
w_values = [3.0]  # Starting point
learning_rate = 0.1
iterations = 20

# Perform gradient descent
for _ in range(iterations):
    current_w = w_values[-1]
    new_w = current_w - learning_rate * gradient(current_w)
    w_values.append(new_w)

# Prepare data for visualization
w_range = np.linspace(-4, 4, 100)
cost_values = cost_function(w_range)

# Plot the cost function and gradient descent path
plt.figure(figsize=(8, 6))
plt.plot(w_range, cost_values, label='Cost Function', color='blue')
plt.scatter(w_values, [cost_function(w) for w in w_values], color='red', label='Gradient Descent Steps')
plt.plot(w_values, [cost_function(w) for w in w_values], linestyle='--', color='red', alpha=0.6)
plt.title("Visualization of Gradient Descent")
plt.xlabel("Parameter (w)")
plt.ylabel("Cost")
plt.legend()
plt.show()
```

```python
# Define the cost function and gradient
def cost_function(w):
    return w**2

def gradient(w):
    return 2 * w

# Simulate gradient descent with different learning rates
learning_rates = [0.01, 0.1, 0.5]
iterations = 20
initial_w = 3.0

plt.figure(figsize=(10, 6))

for lr in learning_rates:
    w_values = [initial_w]
    for _ in range(iterations):
        current_w = w_values[-1]
        new_w = current_w - lr * gradient(current_w)
        w_values.append(new_w)

    # Plot the descent path
    plt.plot(range(iterations + 1), [cost_function(w) for w in w_values], label=f'Learning Rate: {lr}')

plt.title("Impact of Learning Rate on Gradient Descent")
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.legend()
plt.show()
```

```python
# Gradient descent for linear regression
def gradient_descent(X, y, w, b, learning_rate, iterations):
    m = len(y)
    for _ in range(iterations):
        # Compute predictions
        predictions = w * X + b

        # Compute gradients
        dw = (1 / m) * np.sum((predictions - y) * X)
        db = (1 / m) * np.sum(predictions - y)

        # Update parameters
        w -= learning_rate * dw
        b -= learning_rate * db

    return w, b
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Cost function for visualization
def compute_cost(w, b, X, y):
    m = len(y)
    predictions = w * X + b
    return (1 / (2 * m)) * np.sum((predictions - y) ** 2)

# Simulate gradient descent
X = np.array([1, 2, 3, 4, 5])
y = np.array([3, 6, 9, 12, 15])
w, b = 0, 0
learning_rate = 0.01
iterations = 100
costs = []

for _ in range(iterations):
    predictions = w * X + b
    dw = (1 / len(y)) * np.sum((predictions - y) * X)
    db = (1 / len(y)) * np.sum(predictions - y)
    w -= learning_rate * dw
    b -= learning_rate * db
    costs.append(compute_cost(w, b, X, y))

# Plot the cost function over iterations
plt.plot(range(iterations), costs, label="Cost")
plt.title("Cost Function Convergence")
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.legend()
plt.show()
```

```python
def gradient_descent(X, y, learning_rate, iterations):
    m = len(y)  # Number of training examples
    w, b = 0, 0  # Initialize parameters

    for _ in range(iterations):
        # Compute predictions
        predictions = w * X + b

        # Compute gradients
        dw = (1 / m) * np.sum((predictions - y) * X)
        db = (1 / m) * np.sum(predictions - y)

        # Update parameters
        w -= learning_rate * dw
        b -= learning_rate * db

    return w, b
```

```python
# Training data
X = np.array([1, 2, 3, 4, 5])
y = np.array([2.2, 4.1, 6.1, 8.1, 10.2])

# Gradient descent parameters
learning_rate = 0.01
iterations = 1000

# Train the model
w, b = gradient_descent(X, y, learning_rate, iterations)

# Predictions
predictions = w * X + b

# Plot the data and fitted line
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, predictions, color='red', label='Fitted Line')
plt.title('Linear Regression with Gradient Descent')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()

print(f"Learned parameters: w = {w:.2f}, b = {b:.2f}")
```
